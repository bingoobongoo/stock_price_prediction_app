```python
import pandas as pd
import ta
import sqlalchemy
```

```python
%load_ext sql
```

```python
engine = sqlalchemy.create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/stock_price")
%sql postgresql://postgres:postgres@localhost:5432/stock_price
```

```python
tickers = %sql SELECT ticker FROM company
tickers_df = tickers.DataFrame()
tickers_series = tickers_df['ticker']
tickers_series
```

```python
olhc_close_data = %sql SELECT s.company_id, c.ticker, s.date, s.close FROM company as c JOIN stock_price as s USING(company_id) WHERE c.ticker = 'AAL' ORDER BY s.date ASC
```

```python
aal_close_df = olhc_close_data.DataFrame()
aal_close_df
```

```python
rsi = ta.momentum.RSIIndicator(aal_close_df['close'])
rsi.rsi()
```

```python
macd = ta.trend.MACD(aal_close_df['close'])
macd.macd().isna().sum()
```
